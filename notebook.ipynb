```python
mort_df = pd.read_csv("usa_mortality_rates.csv")
```

```python
years_df = pd.DataFrame(mort_df.iloc[0,4:])
years_df = years_df.reset_index().rename(columns = {"index" : "year", 0: "mortality"})
```

```python
fig, ax = plt.subplots()

ax.scatter(years_df.year, 
           years_df.mortality
          )  # save the results of hist

ax.scatter(years_df.year[21], 
           years_df.mortality[21],
           color="red"
          )

ax.set_title("Mortality Rate with Average in Red",loc='left')
ax.set_ylabel("Frequency")
ax.set_xlabel("Mortality")


ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```

```python
df_mortality = pd.read_csv("usa_mortality_rates.csv")
df_mortality.head()
df_rates_only = df_mortality.iloc[0,4:]
avg = df_mortality.iloc[0,4:].mean()
```

```python
fig, ax = plt.subplots()
ax.plot(df_rates_only, lw=6)
plt.axhline(y=avg, color='grey')
ax.text(46,15,'average: 14')
ax.set_title("Mortality Rate, Under 5 (per 1,000 live births)",loc='left')
ax.set_xlabel("Year")
plt.xticks(np.arange(0, 60, step=10))
ax.set_ylabel("Mortality Rate")
left, right = plt.xlim()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```

```python
years_df = years_df.dropna()
years_df.year = years_df.year.astype(int)
years_df.mortality = years_df.mortality.astype(int)
years_df["decade"] = (years_df.year//10)*10
decades_df = years_df[["decade","mortality"]].groupby(by=["decade"]).mean()
```

```python
fig, ax = plt.subplots()

barlist = ax.bar(decades_df.index, 
                 decades_df.mortality,
                 width = 6.0)

barlist[0].set_color('orange')

for rect in barlist.patches:
    rect.set_linewidth(1)
    rect.set_edgecolor('black')

ax.set_title("Average Mortality Rate Per Decade Since 1960", loc="left")
ax.set_ylabel("Mortality")
ax.set_xlabel("Decade")
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)


plt.show()
```

```python
mpl.rcParams['font.size'] = 12

feedback = {'Peers': (85, 91),
            'Culture': (80, 96),
            'Work environment': (76, 75),
            'Leadership': (59, 62),
            'Career development': (49, 33),
            'Rewards & recognition': (41, 45),
            'Perf management': (33, 42)
           }

fig, ax = plt.subplots(figsize=(2,4.8))
plt.rcParams['font.sans-serif'] = "Arial"

# Draw lines and text associated with scores
for c in feedback:
    a,b = feedback[c]
    color = 'dimgrey'
    if c=='Career development':
        color = 'orangered'
    else:
        color = 'dimgrey'
    ax.plot([0,1], [a,b], 'o-', lw=2, c=color)
    ax.text(0-.07, a, f"{a}%", c=color,
        horizontalalignment='right', verticalalignment='center')
    ax.text(1+.07, b, f"{b}%", c=color,
            horizontalalignment='left', verticalalignment='center')
    ax.text(0-.4, a, c, c=color,
            horizontalalignment='right', verticalalignment='center')
    
ax.text(-1.1, 92, 'Survey category | Percent favorable', color='dimgrey', size='10')

# Make the axes look right
ax.set_title("Employee feedback over time", horizontalalignment = 'right', size='15')
ax.spines['bottom'].set_bounds(0, 1)
ax.set_xticks([0,1])
ax.set_xticklabels(['2014','2015'], color='dimgrey')
ax.set_yticks([])
ax.set_ylim(ymin=15)
ax.text(-.11, 2, 'Survey year', color='dimgrey')
ax.text(-1.5, -8, "Figure 1: Slopegraph from “Storytelling with data”", 
        family='serif', name='Times New Roman', size='14')

# Only show the bottom axis
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_linewidth(.5)
plt.show()
```
